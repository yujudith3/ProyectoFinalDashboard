```python
## obtener nombre de archivo cargado ##
file_name = list(uploaded.keys())[0]
```

```python
## dataframe pandas ##
jalisco_pea = pd.read_csv(file_name)
```

```python
print(jalisco_pea.head())
```

```python
## funcion para filtrar datos de columna Poblacion Total POBTOT##

def filtrodatos_pob(jalisco_pea, valor_minimo, valor_maximo):
  datosfiltrados_pob = jalisco_pea[(jalisco_pea['POBTOT']>=valor_minimo)&(jalisco_pea['POBTOT']<=valor_maximo)]
  return datosfiltrados_pob
```

```python
# crear widgets

#Sliders para poner limites en la columna Poblacion Total (POBTOT)
min_slider_pob = pnw.IntSlider(name='valor Minimo', start=0, end=100, step=1, value=0)
max_slider_pob = pnw.IntSlider(name='valor Maximo', start=0, end=100, step=1, value=100)

#Sliders para poner limites en la columna Poblacion Economicamente Activa (PEA)

min_slider_pea = pnw.IntSlider(name='valor Minimo', start=0, end=100, step=1, value=0)
max_slider_pea = pnw.IntSlider(name='valor Maximo', start=0, end=100, step=1, value=100)
```

```python
## visualizacion de sliders ##

pn.Column(min_slider_pob, max_slider_pob).servable()
```

```python
##funcion para filtrar datos de columna Poblacion Economicamente Activa PEA ##

def filtrodatos_pea(jalisco_pea, valor_minimo, valor_maximo):
  datosfiltrados_pea = jalisco_pea[(jalisco_pea['PEA']>=valor_minimo)&(jalisco_pea['PEA']<=valor_maximo)]
  return datosfiltrados_pea
```

```python
## visualizacion de sliders ##

pn.Column(min_slider_pea, max_slider_pea).servable()
```

```python
## definir funcion de actualizacion ##
@pn.depends(min_slider_pob, max_slider_pob)
def actualizardatos_pob(valor_minimo, valor_maximo):
  valor_minimo=min_slider_pob.value
  valor_maximo=max_slider_pob.value
  datosfiltrados_pob=filtrodatos_pob(jalisco_pea, valor_minimo, valor_maximo)
  datosfiltrados_pob.to_csv('datos_filtrados.csv', index=False)
  print("los datos filtrados han sido exportados a 'datos_filtrados.csv'")
```

```python
## definir funcion de actualizacion ##
@pn.depends(min_slider_pea, max_slider_pea)
def actualizardatos_pea(valor_minimo, valor_maximo):
  valor_minimo=min_slider_pea.value
  valor_maximo=max_slider_pea.value
  datosfiltrados_pea=filtrodatos_pea(jalisco_pea, valor_minimo, valor_maximo)
  datosfiltrados_pea.to_csv('datos_filtrados.csv', index=False)
  print("los datos filtrados han sido exportados a 'datos_filtrados.csv'")
```

```python
actualizardatos_pob(min_slider_pob, max_slider_pob)
```

```python
actualizardatos_pea(min_slider_pea, max_slider_pea)
```
